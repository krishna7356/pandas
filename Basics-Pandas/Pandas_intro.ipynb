## Subsetting Data

The process of picking a set of desired rows and columns from a data frame .

```python
# Subsetting Columns
df = df[['title', 'description', 'publishTime',
         'duration_seconds', 'viewCount', 'likeCount',
         'commentCount']]
```

```python
df.shape     # number of columns change to selected columns
```

```python
# Subsetting using loc
df_subset1 = df.loc[df['viewCount'] > 1_000_000]    # sorting the rows based on a condtion(view count >1000000) 
```

##### subsetting the same condition using query

```python
df_subset2 = df.query('viewCount > 1000000')
```

```python
df_subset2 == df_subset1  # both files are same 
```

```python
df = df.loc[~df['likeCount'].isna()]   # removing Nan values from data frame
df
```

```python
df.dtypes
```

## Casting dtypes
covert x varible from one to another( float, integer or string (String should be a whole number) variable)

```python
df['viewCount'] = df['viewCount'].astype('int')  #convert data type float64 to int
df['likeCount'] = df['likeCount'].astype('int')
```

```python
df['publishTime'] = pd.to_datetime(df['publishTime'])  # convert  the  datatype into datetime
```

```python
df['likeCount'] = pd.to_numeric(df['likeCount'].astype('str')) #Convert argument to a numeric type
```

```python
df.dtypes
```

## Creating new column

```python
df['like_to_view_ratio'] = df['likeCount'] / df['viewCount']  # create new column like_to_view_ratio 
```
